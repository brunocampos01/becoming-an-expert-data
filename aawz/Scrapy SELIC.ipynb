## Removing Multple spaces from within string

```python
def remove_multiple_spaces(string):
    """
    Se é String, então faz remoção de espaços em branco 
    replace: / por espaço em branco
    replace: , por . para facilitar a conversão de tipos
    """
    if type(string)==str:
        return ' '.join(string.split()).replace('/',' ').replace(',','.')
    return string
```

```python
response = requests.get("https://www.bcb.gov.br/pec/copom/port/taxaselic.asp", timeout=60)
content = BeautifulSoup(response.content, "html.parser")
table = content.find_all("tr")
HTML(str(table))
```

```python
# neste dataframe vou armazenar a data e a taxa SELIC
df_selic = pd.DataFrame(columns=['Ano', 'Mes', 'Taxa SELIC'])

for row in table:
    text = row.text
    text = remove_multiple_spaces(text)
    
    # insere dados diferentes a cada linha
    text_dados = text.split(sep=" ")
    
    # convert list to String
    ano = ''.join(text_dados[3:4])
    mes = ''.join(text_dados[2:3])
    taxa = ''.join(text_dados[13:])
                
    # variável p organizar a inserção no dataframe
    dados = pd.DataFrame([[ano, mes, taxa]], columns=['Ano', 'Mes', 'Taxa SELIC'])    
    df_selic = df_selic.append(dados)

# remove rows with Strings
df_selic = df_selic[3:]
```

```python
df_selic.Ano = pd.to_numeric(df_selic.Ano, downcast='float', errors='coerence')
df_selic.Mes = pd.to_numeric(df_selic.Mes, downcast='float', errors='coerence')
df_selic['Taxa SELIC'] = pd.to_numeric(df_selic['Taxa SELIC'], downcast='float', errors='coerence')

df_selic.info()
```
